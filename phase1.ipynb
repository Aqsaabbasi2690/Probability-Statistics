## Most Commmon Value

```python
import numpy as np
sizes = np.array([22, 23, 29, 32, 39, 42, 42, 42, 42, 42, 42, 42, 43, 44, 46, 51, 51, 55, 55])

sizes.mean()
```

```python
from collections import Counter
cnt = Counter()

for size in sizes:
    cnt[size] += 1

cnt.most_common()    # index 0 is the most common i.e. the mode
```

 Alternative method:use the scipy library

scipy is scientific computational library for python

```python

from scipy import stats
stats.mode(sizes) 
# index o has value, idex 1 has count
 #mode is essentially another type of avarge(specific name is mean),mode count most common values
# when we want only mode we can write stats.mode(sizes)[0]

# gender.mode() pnadas mode not give us frequency
# stats.mode(gender) stats mode us value and frequency
```
