## Continuous Distribution

# Uniform Distribution

```python
from scipy.stats import uniform
import seaborn as sns
```

```python
n = 10000
# # n ** 10  # This line is commented out in the image and seems to be a note or an alternative
start = 10
width = 20

data_uniform = uniform.rvs(size=n, loc=start, scale=width)
ax = sns.histplot(data_uniform, # Changed from sns.distplot to sns.histplot for histogram
                  bins=100,
                  kde=False,
                 )
ax.set(xlabel='Uniform Distribution', ylabel='Frequency')
```

## Exponential Distributions

```python
# Exponential Distribution
from scipy.stats import expon

data_expon = expon.rvs(scale=1, loc=0, size=1000)
ax = sns.histplot(data_expon, # Changed from sns.distplot to sns.histplot for histogram
                  bins=100,
                  kde=True,
                 )
# # plt.xlim(-10, 10) # This line is commented out in the image
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')

```

## Beta Distribution


```python
from scipy.stats import beta

data_beta = beta.rvs(a=2.31, b=0.627, size=1000)
ax = sns.histplot(data_beta, # Changed from sns.distplot to sns.histplot for histogram
                  bins=100,
                  kde=True,
                 )
# # plt.xlim(-10, 10) # This line is commented out in the image
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

```python
import numpy as np
np.random.seed(1337)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
plt.style.use('ggplot')

num_bins = 50 # number of bins in each dimension Try: 5, 20, 50, 500
num_samples = 10000000

data = 100 + 10 * np.random.randn(num_samples, 2) # generate 100000 2-d random data points
joint_probs, edges = np.histogramdd(data, bins=num_bins)

edges = np.array(edges)[:,1:]
```

## Dependence in joints

```python
import numpy as np

num_bins = 50
num_samples = 10000

# Don't have to understand exactly how these values are being generated
mean = [0, 0]
cov = [[1, 0.8],
       [0.8, 1.0]] # diagonal covariance

data = 100 + 10 * np.random.multivariate_normal(mean, cov, num_samples)
joint_probs, edges = np.histogramdd(data, bins=num_bins)

edges = np.array(edges)[:,1:]
print(joint_probs)
```

```python
cut_point = 14 # try 9, 2 and 14

X_vals = np.array(joint_probs[cut_point])
X_vals /= X_vals.sum()

Y_vals = np.array([row[cut_point] for row in joint_probs])
Y_vals /= Y_vals.sum()

print("Cut X and Y on:", cut_point)
```

```python
plt.bar(edges[0], X_vals) #
plt.ylabel('Frequency')
plt.xlabel('Num');
plt.title("Distribution of X_vals")
plt.show()
```

```python
X = edges[0]
Y = edges[1]
X, Y = np.meshgrid(X, Y)

Z = joint_probs / joint_probs.sum()
```

```python
fig = plt.figure(figsize=(9, 6))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')

ax.view_init(60, 75)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
fig,ax = plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Contour Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```
